#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
employee.groupby(['Department']).agg({'Salary':np.mean}).reset_index()
```

```python
departments_meansalary = employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = ['mean']).reset_index()
departments_meansalary.columns = ['Department', 'Mean salary']
departments_meansalary
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# Custom function to compute the mean after removing the maximum value
def mean_without_max(x):
    if x.mean() == x.max():
        return x.mean()
    else:
        return x[x != x.max()].mean()

# Create pivot table using the custom function
dpt_mean_without_max = employee.pivot_table(index='Department', values='Salary', aggfunc=mean_without_max).reset_index()

dpt_mean_without_max.columns = ['Department', 'Mean of salary(without max)']
dpt_mean_without_max
```
